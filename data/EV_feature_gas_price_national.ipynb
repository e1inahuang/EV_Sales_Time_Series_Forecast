```python
import requests
import pandas as pd

# EIA API Key
api_key = "b2eKxsosbujhRxJSXzafNBCiKynhoBqaAGLQlNQY"

# API endpoint updated to match the structure from the latest screenshot
url = (
    "https://api.eia.gov/v2/petroleum/pri/gnd/data/"
    "?frequency=weekly&data[0]=value"
    "&facets[duoarea][]=NUS"
    "&facets[product][]=EPM0"
    "&start=2014-01-01&end=2024-09-30"
    "&sort[0][column]=period&sort[0][direction]=desc"
    "&offset=0&length=5000"
    f"&api_key={api_key}"
)

# API request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()  # Get the JSON content
    prices = data.get('response', {}).get('data', [])  # Extract the data

    if prices:
        # Convert the data to a pandas DataFrame
        df_us_product = pd.json_normalize(prices)
        df_us_product  # Display the dataframe
    else:
        print("No data available.")
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```
