```python
import requests
import pandas as pd

# EIA API Key
api_key = "b2eKxsosbujhRxJSXzafNBCiKynhoBqaAGLQlNQY"

# API endpoint following the structure in your screenshot
url = (
    "https://api.eia.gov/v2/petroleum/pri/refmg/data/"
    "?frequency=monthly&data[0]=value"
    "&start=2014-01&sort[0][column]=period"
    "&sort[0][direction]=desc&offset=0&length=5000"
    f"&api_key={api_key}"
)

# API request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()  # Get the JSON content
    prices = data.get('response', {}).get('data', [])  # Extract the data

    if prices:
        # Convert the data to a pandas DataFrame
        df_petro2 = pd.json_normalize(prices)

else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
df_gas_pa = df_petro2[df_petro2["area-name"] == "USA-PA"]
```

```python
# Drop rows where the 'value' column has None or NaN values
df_gas_pa_notna= df_gas_pa.dropna(subset=['value'])

# Display the cleaned DataFrame
df_gas_pa_notna
```

```python
# Export the cleaned DataFrame to a CSV file
csv_file_path = "/content/drive/MyDrive/Time series/df_gas_pa_notna.csv"
df_gas_pa_notna.to_csv(csv_file_path, index=False)

print(f"Cleaned DataFrame exported to CSV at {csv_file_path}")
```
